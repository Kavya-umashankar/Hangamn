```python
import requests
from bs4 import BeautifulSoup
#we need 
header={'Useragent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36'}
url="https://economictimes.indiatimes.com/news/politics-and-nation/coronavirus-crisis-heres-total-number-of-confirmed-cases-in-india-as-per-health-ministry/articleshow/74589499.cms"
response= requests.get(url, headers=header)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
stat_table = soup.find_all('table')
```

```python
len(stat_table)
type(stat_table)
```

```python
stat_table=stat_table[0]
```

```python
data=[]
for row in stat_table.find_all('tr'):
    for cell in row.find_all('td'):
        data +=[cell.text]
        #print(cell.text)
print(data)
length=len(data)
```

0,4-name of states
1,5-total confiemed cases(indian)
2,6-total confirmed cases(foregin)
3,7-cured
4,8-death

```python
import pandas as pd
Name_of_State_or_UT = []
Total_Confirmed_cases=[]
Cured =[]
Death = []
headings=data[:4]
Total_India = data[-5:]
for i in range(4,length-4,4):
    Name_of_State_or_UT +=[data[i]]
    Total_Confirmed_cases += [data[i+1]]
    Cured += [data[i+2]]
    Death += [data[i+3]]
print(headings)
print(Total_India)    
```

```python
import pandas as pd
df = pd.DataFrame(list(zip(Name_of_State_or_UT, Total_Confirmed_cases,Cured,Death)), 
               columns =headings)
df
```
